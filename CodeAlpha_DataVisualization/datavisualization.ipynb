```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample dataset
data = {
    "Student": ["A", "B", "C", "D", "E"],
    "Marks": [65, 70, 80, 75, 90]
}

df = pd.DataFrame(data)

# Bar Chart
plt.bar(df["Student"], df["Marks"])
plt.xlabel("Student")
plt.ylabel("Marks")
plt.title("Student Marks Comparison")
plt.show()
```
